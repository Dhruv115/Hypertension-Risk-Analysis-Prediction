# Hypertension Risk Analysis & Prediction
- Focuses on predicting hypertension risk using health and lifestyle factors.
- Dataset includes BP history, medication, family history, exercise, and smoking status.
- Preprocessing involves converting categorical data into numeric form.
- Machine learning models (Logistic Regression, Decision Tree, Random Forest, SVM) are applied.
- Performance evaluated using accuracy, precision, F1-score, and ROC-AUC.
- Goal: Identify key risk factors and build a reliable prediction system for early diagnosis.

# Changing the Object columns to numeric

```python
replace_map={'BP_History': {'Normal': 0, 'Hypertension': 1, 'Prehypertension': 2}, 
             'Medication': {'None': 0, 'ACE Inhibitor': 1, 'Other': 2, 'Beta Blocker': 3, 'Diuretic': 4}, 
             'Family_History': {'Yes': 0, 'No': 1}, 
             'Exercise_Level': {'Low': 0, 'Moderate': 1, 'High': 2}, 
             'Smoking_Status': {'Non-Smoker': 0, 'Smoker': 1}, 
             'Has_Hypertension': {'Yes': 0, 'No': 1}
            }
df.replace(replace_map, inplace=True)
```

```python
df.info()
df.head()
```

# Correlation Matrix

```python
corr_mat = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_mat, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df.isnull().sum()
```

```python
df["Medication"]=df["Medication"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#Features selection
x=df.drop(columns=['Has_Hypertension'])
y=df['Has_Hypertension']
```

```python
#data division into training
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#model preparation
model=LogisticRegression() #model will learn the algorithm not the data 

#model fitting
model.fit(x_train,y_train) #now the model will learn about the data
```

```python
#Prediction From Model
y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy = ",accuracy)

conf_mat=confusion_matrix(y_test,y_pred)
print("\nConfussion Matrix = \n",conf_mat)

class_report=classification_report(y_test,y_pred)
print("\nClassification Report = \n",class_report)
```

#### The accuracy with Logistic Regression is 73 %

## Now using SVM for Predictive Analysis

```python
#Features selection
xx=df.drop(columns=['Has_Hypertension'])
yy=df['Has_Hypertension']
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(xx,yy,test_size=0.3,random_state=42)
```

```python
#feature Scaling
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(X_train)
x_test_scaled=scaler.transform(X_test)
```

```python
svm_model=SVC(kernel='rbf',class_weight='balanced')
svm_model.fit(x_train_scaled,Y_train)
```

```python
Y_pred=svm_model.predict(x_test_scaled)
```

```python
accuracy=accuracy_score(Y_test,Y_pred)
print("Accuracy = ",accuracy)

conf_mat=confusion_matrix(Y_test,Y_pred)
print("\nConfussion Matrix = \n",conf_mat)

class_report=classification_report(Y_test,Y_pred)
print("\nClassification Report = \n",class_report)
```
